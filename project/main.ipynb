```python
from DataProcessor import DataProcessor
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set(style="whitegrid")

crop_path='datasets/worldwide_crop_production.csv'
emission_path='datasets/historical_emissions.csv'
merged_data = DataProcessor(emission_path, crop_path)
merged_data= merged_data.merge_crop_emmision()




```

```python
merged_data.describe()
```

```python
# Plot 1: Time series plot of GHG emissions by type
df = merged_data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['All GHG'], label='All GHG', marker='o')
plt.plot(df.index, df['CH4'], label='CH4', marker='o')
plt.plot(df.index, df['N2O'], label='N2O', marker='o')
plt.xlabel('Year')
plt.ylabel('Emissions (in million tonnes)')
plt.title('Time Series of GHG Emissions in China')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 2: Time series plot of crop production by type
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Maize'], label='Maize', marker='o')
plt.plot(df.index, df['Rice'], label='Rice', marker='o')
plt.plot(df.index, df['Soybean'], label='Soybean', marker='o')
plt.plot(df.index, df['Wheat'], label='Wheat', marker='o')
plt.xlabel('Year')
plt.ylabel('Production (in thousand tonnes)')
plt.title('Time Series of Crop Production in China')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 3: Bar plot of total crop production over the years
plt.figure(figsize=(12, 6))
sns.barplot(x=df.index.year, y=df['Total Crop Production'], palette='viridis')
plt.xlabel('Year')
plt.ylabel('Total Crop Production (in thousand tonnes)')
plt.title('Total Crop Production in China (1990-1993)')
plt.xticks(rotation=45)
plt.show()
```
